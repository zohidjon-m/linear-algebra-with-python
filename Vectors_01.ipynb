```python
class Vector:
    def __init__(self, *components):
        self.components = components

    def __str__(self):
        return f"Vector{self.components}"

    def __repr__(self):
        return self.__str__()
```

```python
import matplotlib.pyplot as plt
vector_1 = np.array([4,2])
vector_2 = np.array([-1, 2])
vector_sum = vector_1 + vector_2

plt.figure()
plt.quiver(0, 0, vector_1[0], vector_1[1], angles='xy', scale_units='xy', scale = 1, color = 'r', label = '(4,2)')
plt.quiver(0, 0, vector_2[0], vector_2[1], angles='xy', scale_units='xy', scale = 1, color = 'g', label = '(-1,2)')
plt.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale = 1, color = 'b', label = ' Sum (3,4)')

plt.xlim(-2,5)
plt.ylim(-1, 5)
plt.grid(True)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Plot')

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a= np.array([2,1])
b= np.array([1,2])
sum_ab = a+b
diff_ab = a - b

print("Vector a:", a)
print("Vector b:", b)
print("Sum of a and b:", sum_ab)
print("Difference of a and b:", diff_ab)

plt.quiver(0,0, a[0], a[1], color = "blue", angles='xy', scale_units='xy', scale=1, label='a')
plt.quiver(0,0, b[0], b[1], color='red', angles='xy', scale_units='xy',  scale=1, label = 'b')
plt.quiver(0,0, sum_ab[0], sum_ab[1], color = 'green', angles='xy', scale_units='xy', scale=1, label='a+b')
plt.quiver(0,0, diff_ab[0], diff_ab[1], color="purple", angles='xy', scale_units="xy", scale=1, label='a-b')
plt.grid(True)
plt.legend()
plt.xlim(-1,4)
plt.ylim(-1,4)
```

```python
#Vector Scaling
v = np.array([1.5, 1.5])
c=2
scaled_v = c*v
print("Scaled Vector: ", scaled_v)
```

```python
#Dot product and angle
#define vecotrs

u= np.array([1,0])
v = np.array([1,1])

#Compute dot product
dot_product = np.dot(u,v)

#compute angle in decrees
angle = np.arccos(dot_product / (np.linalg.norm(u)* np.linalg.norm(v))) * 180 / np.pi

#print with degree symbol
print(f"Dot Product: {dot_product}. Angle:{angle: .1f}\u00b0 ")
```

```python
#Cross Product
u = np.array([1,2,0])
v= np.array([0,1,1])
cross_product = np.cross(u,v)
print("Cross Product: ", cross_product)
```

```python
#Zero Vector
v = np.array([3, -1, 2])
zero = np.zeros_like(v)
print("v+zero: ", v+zero)
print("v*zero: ", v*zero)
```

```python
#Unit vector
v = np.array([1,1,1,1])
unit_v = v/ np.linalg.norm(v)
print(unit_v)
```

```python
#standard basis vectors
n=3
basis = np.eye(n)
print("Basis vecotrs: ")
print("e_1:", basis[0])
print("e_2:", basis[1])
print("e_3:", basis[2])
```

```python
#polynomial as vectors
#coefficients of p(x) = -x^2 + 3x + 1
coeffs = np.array([-1,3,1])

#Generate points for plotting
x = np.linspace(0,3,100)
y = np.polyval(coeffs, x)

#plot the polynomial curve
plt.plot(x,y, label='$p(x) = -x^2 + 3x + 1$', color = "magenta")
plt.grid(True)
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
```
